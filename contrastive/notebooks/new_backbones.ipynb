```python
class CharClass():
    def __init__(self, level):
        self.level = level

class Fighter(CharClass):
    def __init__(self, level, sword):
        super(Fighter, self).__init__(level=level)
        self.sword = sword
        print("fighter")

    def attack(self):
        return 'KLUNK'


class Wizard(CharClass):
    def __init__(self, level, hat):
        super(Wizard, self).__init__(level=level)
        self.hat = hat
        print("wizard")

    def attack(self):
        return 'WIZZZZZ'


class Character(): #?
    def __init__(self, char_class, level=1, sword=None, hat=None):
        if char_class == 'fighter':
            self.classe = Fighter(level=level, sword=sword)
        else:
            self.classe = Wizard(level=level, hat=hat)
        # should inherit from Fighter/Wizard depending on the char_class arg
        print("New character :D")
    
    def attack(self):
        return self.classe.attack()
```

```python
Dumbledore = Character('wizard', level=20, hat='Choipeau')

Dumbledore.attack()
```
