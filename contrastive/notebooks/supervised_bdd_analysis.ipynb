```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# schiz

```python
df = pd.read_csv('/neurospin/dico/agaudin/Runs/09_new_repo/Output/supervised/schiz/bdd_schiz.csv', index_col=0)
df.datasets = df.datasets.astype(str)

corresp_dico = {"('cingulate_schiz_left',)": 'left',
                "('cingulate_schiz',)": 'right',
                "('cingulate_schiz', 'cingulate_schiz_left')": 'both'}

df['side'] = df.datasets.map(corresp_dico)
df
```

```python
df[df.side == 'both']
```

```python
cols = ['train_auc', 'val_auc', 'test_auc', 'test_intra_auc', 'best_train_auc', 'best_val_auc', 'best_test_auc', 'best_test_intra_auc']

fig = plt.figure(figsize=(32,8))

for i, col in enumerate(cols):
    ax = fig.add_subplot(1, len(cols), i+1)
    sns.violinplot(df, x='side', y=col, kind='violin', ax=ax)
    plt.title(col)
```

```python
cols = ['train_auc', 'val_auc', 'test_auc', 'test_intra_auc', 'best_train_auc', 'best_val_auc', 'best_test_auc', 'best_test_intra_auc']

plt.figure(figsize=(32,8))

for i, col in enumerate(cols):
    plt.subplot(1, len(cols), i+1)
    ax = plt.gca()
    ax.set_ylim(0,1.05)
    sns.barplot(df, x='side', y=col)
    sns.stripplot(df, x='side', y=col, color='black')
    plt.title(col)
```
