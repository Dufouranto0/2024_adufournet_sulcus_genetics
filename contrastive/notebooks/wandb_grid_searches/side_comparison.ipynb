```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def gs_crit_full(df):
    overfit = np.abs(df.train_auc-df.val_auc) + np.abs(df.train_auc-df.test_auc) + \
              np.abs(df.train_auc-df.test_intra_auc)
    best_perf = 1-df.best_train_auc + 1-df.best_val_auc + 1-df.best_test_intra_auc + \
                1-df.best_test_auc
    
    return overfit + 2*best_perf
```

```python
palette = {"('STs_schiz_L_strat_bis',)": 'C2',
           "('STs_schiz_R_strat_bis',)": 'C0',
           "('STs_schiz_R_strat_bis', 'STs_schiz_L_strat_bis')": 'C1'}
```

# S.T.s.

```python
bdd_sts = pd.read_csv("/neurospin/dico/agaudin/Runs/09_new_repo/Output/grid_searches/step2/STs/bdd_STs.csv", index_col=0)
bdd_sts
```

```python
plt.figure(figsize=(32,8))

plt.subplot(141)
sns.barplot(bdd_sts, y='best_train_auc', x='datasets', palette=palette)
plt.xticks(rotation=90)

plt.subplot(142)
sns.barplot(bdd_sts, y='best_val_auc', x='datasets', palette=palette)
plt.xticks(rotation=90)

plt.subplot(143)
sns.barplot(bdd_sts, y='best_test_intra_auc', x='datasets', palette=palette)
plt.xticks(rotation=90)

plt.subplot(144)
sns.barplot(bdd_sts, y='best_test_auc', x='datasets', palette=palette)
plt.xticks(rotation=90)
plt.show()
```

```python
bdd_sts['criterion'] = gs_crit_full(bdd_sts)
bdd_sts = bdd_sts.sort_values(by='criterion')

plt.figure(figsize=(16,6))

plt.subplot(121)
sns.barplot(bdd_sts, y='criterion', x='model_path', hue='datasets', palette=palette)
plt.xticks([], [])

plt.subplot(122)
sns.barplot(bdd_sts, y='criterion', x='datasets', palette=palette)
plt.xticks([], [])

plt.show()

# more often overfitting when both sides
```

```python
bdd_sts['min_best_perf'] = bdd_sts[['best_train_auc', 'best_val_auc', 'best_test_intra_auc', 'best_test_auc']].apply(min, axis=1)

bdd_sts = bdd_sts.sort_values(by='min_best_perf', ascending=False)

plt.figure(figsize=(16,6))

plt.subplot(121)
sns.barplot(bdd_sts, y='min_best_perf', x='model_path', hue='datasets', palette=palette)
plt.xticks([], [])

plt.subplot(122)
sns.barplot(bdd_sts, y='min_best_perf', x='datasets', palette=palette)
plt.xticks([], [])

plt.show()

bdd_sts[:10]
```
